```python
import math
from matplotlib import pyplot as plt

-0.5*math.log(0.5)-0.5*math.log(0.5)
# [1, 1, 1]
print(-0.333*math.log(0.333)-0.333*math.log(0.333)-0.333*math.log(0.333))
# [2, 1, 0]
-0.667*math.log(0.667)-0.333*math.log(0.333)

def entropy(n):
    return -n * (1/n)*math.log(1/n)

nvalues = range(1, 1000)
evalues = [entropy(n) for n in nvalues]
plt.plot(nvalues, evalues)
```
